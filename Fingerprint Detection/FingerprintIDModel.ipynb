```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential, save_model, load_model
```

```
modelPath = '/content/SubjectIDModel.hdf5'
model = load_model(modelPath, compile = True)
```

```
img_path = '/content/datasets_38300_58521_SOCOFing_Real_108__M_Right_ring_finger.dib'
img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
img_size = 96
img_resize = cv2.resize(img_array, (img_size, img_size))
X_test = np.array(img_resize).reshape(-1, img_size, img_size, 1)
X_test = X_test / 255.0
plt.imshow(X_test.reshape((96, 96)), cmap ='gray')
y_SubjectID_pred = model.predict(X_test)
Id_pred = np.argmax(y_SubjectID_pred)
print('This fingerprint is of id:', Id_pred+1) # +1 because model is predicting one value less
```
